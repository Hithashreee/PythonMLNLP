```python
X=df.drop(columns='Outcome',axis=1)
X.head()
```

```python
y=df['Outcome']
y.head()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
Log=LogisticRegression()
```

```python
Log
```

```python
Log.fit(xtrain,ytrain)
```

```python
pred=Log.predict(xtest)
pred
```

```python
accuracy=accuracy_score(ytest,pred)
accuracy
```

```python
import gradio as gr
```

```python
def predict(name,Pregnancies,Glucose,BloodPressure,SkinThickness,Insulin,BMI,Age,DiabetesPedigreeFunction):
  model=joblib.load('/content/logR.pkl')
  dtmodel=joblib.load('dt.pkl')

  ar=np.array([Pregnancies,Glucose,BloodPressure,SkinThickness,Insulin,BMI,Age,DiabetesPedigreeFunction]).reshape(1,-1)
  anslr=model.predict(ar)
  ansdt=dtmodel.predict(ar)
  if(int(ansdt)==1):
    dtans=f'{name} has diabetes'
  else:
    dtans=f'{name} is normal'
  if(int(anslr)==1):
    lrans=f'{name} has diabetes'
  else:
    lrans=f'{name} is normal'
  return dtans,lrans
```

```python
iface=gr.Interface(
  fn=predict,
  inputs=[gr.Text(label='Enter Patient Name'),gr.Number(label='Enter Pregnancies level '),gr.Number(label='Enter Glucose Level'),gr.Number(label='Enter BloodPressure level'),gr.Number(label='Enter SkinThickness level'),gr.Number(label='Enter Insulin level'),gr.Number(label='Enter BMI'),gr.Number(label='Enter DiabetesPedigreeFunction'),gr.Number(label='Enter Age')],
  outputs=[gr.Text(),gr.Text()],
  title='Diabetes Predictor'
)
iface.launch()
```
